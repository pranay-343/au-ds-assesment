## Fetch Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
au_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science assessment for Aunalytics/au_train.csv')
```

```python
au_train.head()
```

```python
au_train = au_train.replace(' ?',np.nan)
au_train_no_missing = au_train.dropna()
```

## Visualization

Identifying Categorical and Numerical Data Keys

```python
categorical = au_train_no_missing.select_dtypes(include=['object']).keys()
categorical
```

```python
Numerical = au_train_no_missing.select_dtypes(include=['int64']).keys()
Numerical
```

Plots of Categorical Data

Number of records for both income class - By this graph we can identify that the income class level less than USD 50K is more than 3 times of those above USD 50K, indicating that the the dataset is somewhat skewed.

```python
fig = plt.figure(figsize=(10,5))
sns.countplot(x='class',hue='class', data=au_train_no_missing,palette="Set2")
plt.xticks(rotation=45)
plt.xlabel('Class')
plt.ylabel("No. of People")
plt.rcParams.update({'font.size': 12})
plt.title('Individual classified inaccordance with their class', fontsize=12)
plt.tight_layout()
plt.show()
```

Working classes classified as per their income class

Private sector has the largest count for both of the income classes. 
This indicates that the given sample of individuals is more inclied towards working for private sector rather than being self employed or working for the government.

```python
fig = plt.figure(figsize=(10,5))
sns.countplot(x='workclass',hue='class', data=au_train_no_missing,palette="Set2",order = au_train_no_missing['workclass'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Working class')
plt.ylabel("No. of People")
plt.rcParams.update({'font.size': 12})
plt.title('Count of Individuals as per Workclasses', fontsize=12)
plt.tight_layout()
plt.show()
```

Income of individuals according to their highest level of education

Here we can identify that more than 55% of master's students have more than 50K income whereas HS-grad have the highest percent of individulas who earn less than $50k which is equal to 84%.



```python
fig = plt.figure(figsize=(10,5))
sns.countplot(x='education',hue='class', data=au_train_no_missing,palette="Set2",order = au_train_no_missing['education'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Education')
plt.ylabel("No. of People")
plt.rcParams.update({'font.size': 12})
plt.title('Income classified as per highest level of education', fontsize=12)
plt.tight_layout()
plt.show()
```

Classification of income class with respect to marital status

Distribution plot for age w.r.t class

For income level greater than 50K USD the density plot is similar to the gaussian density plot

```python
fig = plt.figure(figsize=(10,5))
sns.distplot(au_train_no_missing[au_train_no_missing['class'] == ' >50K']['age'], kde_kws={"label": ">50K"})
sns.distplot(au_train_no_missing[au_train_no_missing['class'] == ' <=50K']['age'], kde_kws={"label": "<=50K"})
plt.rcParams.update({'font.size': 12})
plt.legend(loc=0,fontsize=12)
plt.title('Density Distribution Plot of Income Class versus Age ', fontsize=12)
plt.tight_layout()
plt.show()
```

Distribution plot for hours per week w.r.t income class


We can see that most individuals work for 40 hours each week. But it can se seen that a large number individulas working for more than 40 hours have a income higher than $50k.

```python
fig = plt.figure(figsize=(10,5))
sns.distplot(au_train_no_missing[au_train_no_missing['class'] == ' >50K']['hours-per-week'], kde_kws={"label": ">50K"})
sns.distplot(au_train_no_missing[au_train_no_missing['class'] == ' <=50K']['hours-per-week'], kde_kws={"label": "<=50K"})
plt.rcParams.update({'font.size': 12})
plt.title('Density Distribution Plot of Number of Hours Indivisuals Work Each Week', fontsize=12)
plt.legend(loc=0,fontsize=12)
plt.tight_layout()
plt.show()
```

Age vs Class

The box plots show that both income classses have a large number of outliers indicating data is imbalanced. Also person with more experience/age has more income.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='class',y='age',data=au_train_no_missing,palette='Set2')
plt.rcParams.update({'font.size': 12})
plt.xticks(rotation=45)
plt.title('Box Plot for Age versus Income Class', fontsize=12)
plt.tight_layout()
plt.show()
```

Occupatoin vs age

The interquartile range is highest for Priv-house-service indicating this occupation has a more diverse age group.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='occupation',y='age',data=au_train_no_missing,palette='Set2')
plt.rcParams.update({'font.size': 12})
plt.xticks(rotation=45)
plt.title('Box Plot for Occupation versus Age', fontsize=12)
plt.tight_layout()
plt.show()
```

Occupation vs hours per week

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='occupation',y='hours-per-week',data=au_train_no_missing,palette='Set2')
plt.rcParams.update({'font.size': 12})
plt.xticks(rotation=45)
plt.title(' Box Plot for Occupation versus hours worked per week ', fontsize=12)
plt.tight_layout()
plt.show()
```
